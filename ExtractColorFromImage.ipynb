```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
#to determine lowaer and upper hsv for color
def hsv(color):
    if color == "green":
        green = np.uint8([[[0,255,0]]])
        hsv_green =cv2.cvtColor(green , cv2.COLOR_BGR2HSV)
        print("green =  " , hsv_green)
        return hsv_green
    elif color =="blue" :
        blue = np.uint8([[[255,0,0]]])
        hsv_blue =cv2.cvtColor(blue , cv2.COLOR_BGR2HSV)
        print("blue =  " ,hsv_blue)
        return blue
    elif color =="red" :
        red = np.uint8([[[255,0,0]]])
        hsv_red =cv2.cvtColor(red , cv2.COLOR_BGR2HSV)
        print("red =  " , hsv_red)
        return red

color_dict_HSV = {'black': [[180, 255, 30], [0, 0, 0]],
              'white': [[180, 18, 255], [0, 0, 231]],
              'red1': [[180, 255, 255], [159, 50, 70]],
              'red2': [[9, 255, 255], [0, 50, 70]],
              'green': [[89, 255, 255], [36, 50, 70]],
              'blue': [[128, 255, 255], [90, 50, 70]],
              'yellow': [[35, 255, 255], [25, 50, 70]],
              'purple': [[158, 255, 255], [129, 50, 70]],
              'orange': [[24, 255, 255], [10, 50, 70]],
              'gray': [[180, 18, 230], [0, 0, 40]]
                 }       

img = cv2.imread("color.jpg" )
hsv('green')
hsv('blue')
hsv('red')

plt.imshow(img)
plt.show()
```

```python
#detect blue  in image
import cv2
from matplotlib import pyplot as plt
img = cv2.imread("color.jpg" )
#green = hsv("green")
hsv =cv2.cvtColor(img ,cv2.COLOR_BGR2HSV)

#define range for blue color in img
lower_blue = np.array([110,50,50])
upper_blue =np.array([130,255,255])

# Threshold the HSV image to get only blue colors
mask = cv2.inRange(hsv, lower_blue, upper_blue)

# Bitwise-AND mask and original image
res = cv2.bitwise_and(img,img, mask= mask)
imge =cv2.cvtColor(img ,cv2.COLOR_BGR2RGB)

plt.imshow(imge)
plt.show()
maske =cv2.cvtColor(mask ,cv2.COLOR_BGR2RGB)

plt.imshow(maske)
plt.show()
result =cv2.cvtColor(res ,cv2.COLOR_BGR2RGB)
plt.imshow(result)
plt.show()
```
